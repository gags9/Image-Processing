```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("soccer.bmp",0)
plt.imshow(img)
plt.show()
```

```python
kernel=np.ones((5,5),np.float32)/25

#cv2.filter2D() 
#1st argument is source img, 
#2nd argument is ddepth which the range of the output,-1 means same as imput, 
#3rd argument is filter operator 
result=cv2.filter2D(img,-1,kernel)



result=result.astype(np.uint8)

plt.subplot(121)
plt.imshow(img,cmap='gray',vmin=0,vmax=255)

plt.subplot(122)
plt.imshow(result,cmap='gray',vmin=0,vmax=255)



plt.show()
```

```python
#Sharpening
kernel=np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
sharp=cv2.filter2D(img,-1,kernel)

plt.subplot(1,2,1)
plt.imshow(img,cmap='gray',vmin=0,vmax=255)
plt.title('original image')

plt.subplot(1,2,2)
plt.imshow(sharp,cmap='gray',vmin=0,vmax=255)
plt.title("sharpened image")

plt.show()
```

```python
# Edge detection
kernel=np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
edge=cv2.filter2D(img,-1,kernel)

plt.subplot(1,2,1)
plt.imshow(img,cmap='gray',vmin=0,vmax=255)
plt.title('original image')

plt.subplot(1,2,2)
plt.imshow(edge,cmap='gray',vmin=0,vmax=255)
plt.title("edges")

plt.show()
```
