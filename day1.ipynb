Image Processing

Import relevent packages

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("image.jpg")# Reads image and stores data into an numpy array
print(img.shape)#gives output in (heigt, width, channels) 
```

```python
cv2.imshow("Image",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
plt.imshow(img)
plt.show()
```

```python
soccerimg=cv2.imread("soccer.bmp")#openCV original image is bgr format
imgrgb=cv2.cvtColor(soccerimg,cv2.COLOR_BGR2RGB)
fig, axes = plt.subplots(1, 2)
axes[0].imshow(soccerimg)
axes[1].imshow(imgrgb)
plt.show()
```

```python
gray1=cv2.cvtColor(soccerimg,cv2.COLOR_BGR2GRAY)
gray2=cv2.cvtColor(soccerimg,cv2.COLOR_RGB2GRAY)
fig,axes=plt.subplots(1,2)
axes[0].imshow(gray1)
axes[1].imshow(gray2)
```

```python
resized300=cv2.resize(soccerimg,(300,300))
resized200=cv2.resize(soccerimg,(200,200))
resized100=cv2.resize(soccerimg,(100,100))
fig,axes=plt.subplots(2,2)
axes[0,0].imshow(soccerimg)
axes[0,1].imshow(resized300)
axes[1,0].imshow(resized200)
axes[1,1].imshow(resized100)

plt.show()
```

```python
crop=soccerimg[200:400,200:400]
plt.imshow(crop)
plt.show()
cv2.imwrite("outimg.bmp",crop)
```
