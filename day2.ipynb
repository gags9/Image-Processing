```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.cvtColor(cv2.imread("soccer.bmp"),cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
plt.show()
```

```python
hsv=cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
h,s,v=cv2.split(hsv)
fig,axes=plt.subplots(1,4)
axes[0].imshow(h)#range from 0 to 180 for color type
axes[1].imshow(s)#saturation color purity, range from 0 to 255, 0 indicating no color and 255 pure color, 
axes[2].imshow(v)#
axes[3].imshow(hsv)#pyplot uses image as RGB format only
plt.show()
```

```python
lowergreen=(100,20,20)
uppergreen=(130,255,255)
mask=cv2.inRange(hsv,lowergreen,uppergreen)

#clrmask=cv2.cvtColor(mask,cv2.COLOR_HSV2RGB)
plt.imshow(mask)
plt.show()
print(mask)

```
